importations du fichier csv dans pandas

```python
import pandas as pd
import re 
import numpy as np

df = pd.read_csv("referentiel.csv", sep=",",  encoding="utf-8")
df.head()
```

Aficher les information sur les colonnes

```python
df.info()
```

prétraitement du text 

```python
import pandas as pd
import re

def clean_text(text):
    text = text.lower()                     # tout en minuscule
    text = re.sub(r'\s+', ' ', text)        # retirer les espaces multiples
    text = text.strip()                     # retirer espaces début/fin
    return text

df["skill"] = df["skill"].apply(clean_text)

df.head()
```

importé SBERT

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

Transformer du texte → vecteur numérique (embedding)

```python
df["embedding"] = df["skill"].apply(lambda x: model.encode(x))
df.head()
```

Sauvegarder ton référentiel vectorisé

```python
df.to_pickle("referentiel_embeddings.pkl")
```
